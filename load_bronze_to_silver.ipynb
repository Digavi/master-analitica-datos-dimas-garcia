```python
import os
import pandas as pd
```

```python
bronze_path = os.path.join(os.curdir, "Data_Lake","Bronze")
silver_path = os.path.join(os.curdir, "Data_Lake","Silver")
```

```python
bronze_file_name = "atletas_bronze.csv"
silver_file_name = "atletas_silver.csv"
```

```python
bronze_file_path = os.path.join(bronze_path,bronze_file_name)
silver_file_path = os.path.join(silver_path,silver_file_name)
```

```python
bronze_atletas_df  = pd.read_csv(bronze_file_path)
```

```python
bronze_atletas_df.head()
```

```python
#La columna unnamed corresponde a los índices del anterior csv (antes de que Jorge los ensuciara), por lo que procedo a eliminarla
bronze_atletas_eliminar_unnamed_df = bronze_atletas_df.drop(columns=['Unnamed: 0'])
```

```python
bronze_atletas_eliminar_unnamed_df.head()
```

```python
bronze_atletas_eliminar_unnamed_df.info()
```

```python
#Veo que no hay ningun nulo en id, voy a ver que sean todos numeros
bronze_atletas_eliminar_unnamed_df['ID'].value_counts()
```

```python
#bronze_atletas_eliminar_unnamed_df['ID'] = bronze_atletas_eliminar_unnamed_df['ID'].astype('int')
#Me da este error: "invalid literal for int() with base 10: 'nueve'", por lo que se que hay un nueve metido, lo cambio a 9
bronze_atletas_eliminar_unnamed_df['ID'] = bronze_atletas_eliminar_unnamed_df['ID'].replace("nueve",9)
```

```python
#Sigue habiendo un valor que se llama ds
ds_df = bronze_atletas_eliminar_unnamed_df[bronze_atletas_eliminar_unnamed_df['ID'].str.contains('ds', na=False)]
print(ds_df)
#Veo que Mohammed Abbas Sabih es el id ds, voy a ver si hay otra fila suya para ponerle su id
```

```python
Mohammed_df = bronze_atletas_eliminar_unnamed_df[bronze_atletas_eliminar_unnamed_df['Name'].str.contains('Mohammed Abbas Sabih', na=False)]
print(Mohammed_df)
#No veo otra fila en la que aparezca, por lo que voy a añadirle un id cualquiera
```

```python
#Al usar head al principio, vi que no habia un 3 en id, voy a ver si hay algun caso y sino, se lo pongo a Mohammed
ID3_df = bronze_atletas_eliminar_unnamed_df[bronze_atletas_eliminar_unnamed_df['ID'] == '3']
print(ID3_df)
```

```python
#Veo que está vacío, se lo pongo a Mohammed
bronze_atletas_eliminar_unnamed_df['ID'] = bronze_atletas_eliminar_unnamed_df['ID'].replace("ds",3)
```

```python
#Vuelvo a intentar transformar en numero
bronze_atletas_eliminar_unnamed_df['ID'] = bronze_atletas_eliminar_unnamed_df['ID'].astype('int')
```
