```python
#Creo una funcion para añadir la temporada si tengo los juegos
def añadir_season(fila):
    if pd.isna(fila['Season']) and pd.notna(fila['Games']):
        temporada = fila['Games'].split()
        #Añado condicion de que en 'Games' hay 2 palabras separadas por un espacio
        if len(temporada) > 1:
            return temporada[1]
        else:
            return None
    else:
        return fila['Season']
```

```python
bronze_atletas_season_df['Season'] = bronze_atletas_season_df.apply(añadir_season, axis = 1)
```

```python
#Voy a hacer lo mismo, pero en vez de si esta vacio, si el año es mayor que 2025
def añadir_year(fila):
    if fila['Year'] > 2025 and pd.notna(fila['Games']):
        temporada = fila['Games'].split()
        #Añado condicion de que en 'Games' hay 2 palabras separadas por un espacio
        if len(temporada) > 1:
            return int(temporada[0])
        else:
            return None
    else:
        return fila['Year']

```

```python
bronze_atletas_year_df['Year'] = bronze_atletas_year_df.apply(añadir_year, axis = 1)
```

```python
#Voy a hacer una funcion para añadir los juegos
def añadir_games(fila):
    if pd.isna(fila['Games']) and pd.notna(fila['Season']) and pd.notna(fila['Year']):  #Si no estan los juegos, pero si el año y la estacion
        return f"{fila['Year']} {fila['Season']}"                                       #Entonces compongo los juegos con el año y la estacion
    elif re.match(r'^\d{4}$', fila['Games']) and pd.notna(fila['Season']) and pd.notna(fila['Year']): #Aqui miro que si los juegos aparece solo el año hago lo mismo que antes
        return f"{fila['Year']} {fila['Season']}"
    else:
        return fila['Games']
```

```python
bronze_atletas_games_df['Games'] = bronze_atletas_games_df.apply(añadir_games, axis = 1)
#Aplico la funcion
```
