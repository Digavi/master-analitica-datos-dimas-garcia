```python
import os
import pandas as pd
```

```python
gold_path = os.path.join(os.curdir, "Data_Lake","Gold")
gold_2_path = os.path.join(os.curdir, "Data_Lake","Gold")
```

```python
gold_file_name = "atletas_gold.csv"
gold_2_file_name = "atletas_medallas_gold.csv"
```

```python
gold_file_path = os.path.join(gold_path,gold_file_name)
gold_2_file_path = os.path.join(gold_2_path,gold_2_file_name)
```

```python
gold_atletas_df = pd.read_csv(gold_file_path)
```

```python
gold_atletas_df.head()
```

```python
medallas_por_pais_df = gold_atletas_df.groupby(['Year','Country','Medal']).size().unstack(fill_value=0)
#Agrupo por año, pais y medallas para ver cuantas medallas consiguió cada pais en cada año. El unstack(fill_value=0) es para que aparezca en columnas las medallas por pais y para que los que no tengan medalla, me salga 0
```

```python
print(medallas_por_pais_df)
```

```python
participantes_df = gold_atletas_df.groupby(['Year','Country']).size().rename('Participantes')
#Creo un nuevo data en el que cuento todos los participantes por año y pais
```

```python
participantes_medallas_por_pais_df = medallas_por_pais_df.join(participantes_df,how='outer').fillna(0)
#Junto ambas tablas
```

```python
participantes_medallas_por_pais_df[['Bronze','Silver','Gold']] = participantes_medallas_por_pais_df[['Bronze','Silver','Gold']].astype(int)
#Cambio las columnas de las medallas a enteros para que me salga 4 en vez de 4.0
```

```python
participantes_medallas_por_pais_df['Total_Medallas'] = participantes_medallas_por_pais_df[['Gold','Silver','Bronze']].sum(axis=1)
#Añado una columna con el total de medallas
```

```python
print(participantes_medallas_por_pais_df)
```

```python
#Creo carpeta
os.makedirs(gold_2_path, exist_ok=True)
```

```python
participantes_medallas_por_pais_df.to_csv(gold_2_file_path, index=True)
```
